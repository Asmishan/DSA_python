```python
students = []

def add_student():
    roll_number = input("Enter Roll Number: ")
    name = input("Enter Name: ")
    python_marks = int(input("Enter Marks in Python: "))
    data_science_marks = int(input("Enter Marks in Data Science: "))
    mathematics_marks = int(input("Enter Marks in Mathematics: "))

    student = {
        'Roll Number': roll_number,
        'Name': name,
        'Python': python_marks,
        'Data Science': data_science_marks,
        'Mathematics': mathematics_marks
    }
    students.append(student)
    print("Student added successfully!")
def Search_student_by_rollno():
    roll_number = input("Enter Roll Number to search: ")
    for student in students:
        if student['Roll Number'] == roll_number:
            print("\n")
            print(f"Roll Number: {student['Roll Number']}")
            print(f"Name: {student['Name']}")
            print(f"Python: {student['Python']}, Data Science: {student['Data Science']}, Mathematics: {student['Mathematics']}")
            return
    print("Student not found.")

def display_all_students():
    if not students:
        print("No students to display.")
        return
    for student in students:
        print("\n")
        print(f"Roll Number: {student['Roll Number']}")
        print(f"Name :{student['Name']}")
        print(f"Python: {student['Python']}, Data Science: {student['Data Science']}, Mathematics: {student['Mathematics']}")
    print("--------------------------------------------------")
def find_topper():
    if not students:
        print("No students to find the topper for.")
        return
    topper = students[0]
    for student in students:
        if student['Python'] + student['Data Science'] + student['Mathematics'] > topper['Python'] + topper['Data Science'] + topper['Mathematics']:
            topper = student
    print("\n")
    print(f"Topper: Roll Number: {topper['Roll Number']}")
    print(f"Name: {topper['Name']}")
    print(f"  Python: {topper['Python']}, Data Science: {topper['Data Science']}, Mathematics: {topper['Mathematics']}")
def students_passed():
    if not students:
        print("No students to display.")
        return
    print("Students who passed:")
    for student in students:
        if student['Python'] >= 50 and student['Data Science'] >= 50 and student['Mathematics'] >= 50:
            print("\n")
            print(f"Roll Number: {student['Roll Number']}")
            print(f"Name :{student['Name']}")
            print(f"  Python: {student['Python']}, Data Science: {student['Data Science']}, Mathematics: {student['Mathematics']}")
    print("--------------------------------------------------")
def students_failed_count():
    if not students:
        print("No students to count.")
        return
    count = 0
    for student in students:
        if student['Python'] < 40 or student['Data Science'] < 40 or student['Mathematics'] < 40:
            count += 1
            print("\n")
            print(f"Roll Number: {student['Roll Number']}")
            print(f"Name :{student['Name']}")
            print(f"  Python: {student['Python']}, Data Science: {student['Data Science']}, Mathematics: {student['Mathematics']}")
    print(f"Total number of students failed: {count}")
    print("--------------------------------------------------")
def highest_score():
    if not students:
        print("No students to display.")
        return

    highest_marks = {
        'Python': 0,
        'Data Science': 0,
        'Mathematics': 0
    }
    highest_scorers = {
        'Python': [],
        'Data Science': [],
        'Mathematics': []
    }

    for student in students:
        if student['Python'] > highest_marks['Python']:
            highest_marks['Python'] = student['Python']
            highest_scorers['Python'] = [student]
        elif student['Python'] == highest_marks['Python']:
            highest_scorers['Python'].append(student)

        if student['Data Science'] > highest_marks['Data Science']:
            highest_marks['Data Science'] = student['Data Science']
            highest_scorers['Data Science'] = [student]
        elif student['Data Science'] == highest_marks['Data Science']:
            highest_scorers['Data Science'].append(student)

        if student['Mathematics'] > highest_marks['Mathematics']:
            highest_marks['Mathematics'] = student['Mathematics']
            highest_scorers['Mathematics'] = [student]


    print("Subject-wise Highest Marks:")
    for subject, marks in highest_marks.items():
        print(f"{subject}: {marks}")
        for student in highest_scorers[subject]:
            print("\n")
            print(f"  Roll Number: {student['Roll Number']}, Name: {student['Name']}")
        print()
while True:
    print("\n===== STUDENT MANAGEMENT =====")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Search Student by Roll Number")
    print("4. Find Topper")
    print("5. Display Students who Passed")
    print("6. Count Students Failed")
    print("7. Display Subject-wise Highest Marks")
    print("8. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        add_student()
    elif choice == '2':
        display_all_students()
    elif choice =='3':
        Search_student_by_rollno()
    elif choice =='4':
        find_topper()
    elif choice =='5':
        students_passed()
    elif choice =='6' :
        students_failed_count()
    elif choice  =='7' :
        highest_score()
    elif choice == '8':
        print("-----EXIT-----")
        break
    else:
        print("Invalid choice. Please try again.")
```
